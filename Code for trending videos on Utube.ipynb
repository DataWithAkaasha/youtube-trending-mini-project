```python
# ---------------------------
# STEP 0: Install library (only run once per environment)
# ---------------------------
!pip install google-api-python-client pandas

# ---------------------------
# STEP 1: Import libraries 
# ---------------------------
import pandas as pd
import datetime
from googleapiclient.discovery import build

# ---------------------------
# STEP 2: Setup API
# ---------------------------
API_KEY = # Replace with your key
youtube = build("youtube", "v3", developerKey=API_KEY)

# ---------------------------
# STEP 3: Define countries
# ---------------------------
countries = {
    "Pakistan": "PK",
    "India": "IN",
    "Germany": "DE"
}

# ---------------------------
# STEP 4: Collect data
# ---------------------------
today = datetime.date.today().strftime("%Y-%m-%d")
all_data = []

for country, code in countries.items():
    request = youtube.videos().list(
        part="snippet,statistics",
        chart="mostPopular",
        regionCode=code,
        maxResults=50   # 50 trending videos per country
    )
    response = request.execute()
    
    for item in response["items"]:
        video = {
            "video_id": item["id"],
            "title": item["snippet"]["title"],
            "channel": item["snippet"]["channelTitle"],
            "publish_time": item["snippet"]["publishedAt"],
            "views": item["statistics"].get("viewCount"),
            "likes": item["statistics"].get("likeCount"),
            "comments": item["statistics"].get("commentCount"),
            "country": country,
            "date_collected": today
        }
        all_data.append(video)

# ---------------------------
# STEP 5: Save CSV
# ---------------------------
df = pd.DataFrame(all_data)

# make sure "data" folder exists
import os
os.makedirs("data", exist_ok=True)

filename = f"data/trending_{today}.csv"
df.to_csv(filename, index=False)

print(f" Saved {filename} with {df.shape[0]} rows")
df.head()
```
